### gpt records

```python
import nltk
for index, row in df.iterrows():
    #query = row["question"]
    # records
    reference = row["Answer"] # ground truth
    candidate_400 = row["answer-records-400-OpenAPI"]
    candidate_200 = row["answer-records-200-OpenAPI"]
    candidate_600 = row["answer-records-600-OpenAPI"]
    
    # Tokenize the reference and candidate responses
    reference_tokens = nltk.word_tokenize(reference.lower())
    candidate_tokens_400 = nltk.word_tokenize(candidate_400.lower())
    candidate_tokens_200 = nltk.word_tokenize(candidate_200.lower())
    candidate_tokens_600 = nltk.word_tokenize(candidate_600.lower())

    # Calculate BLEU score
    bleu_score_400 = nltk.translate.bleu_score.sentence_bleu([reference_tokens], candidate_tokens_400)
    bleu_score_200 = nltk.translate.bleu_score.sentence_bleu([reference_tokens], candidate_tokens_200)
    bleu_score_600 = nltk.translate.bleu_score.sentence_bleu([reference_tokens], candidate_tokens_600)
    #print("BLEU Score:", bleu_score)
    
    df.at[index,"bleu_score_gpt_record_400"] = bleu_score_400
    df.at[index,"bleu_score_gpt_record_200"] = bleu_score_200
    df.at[index,"bleu_score_gpt_record_600"] = bleu_score_600
#print("BLEU Score:", bleu_score)
```
