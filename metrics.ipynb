```python
df = pd.read_csv("ground_truth.csv",encoding='latin-1')
#df.columns = ["question","ground_truth","rag_answers_400","llm_answers_400"]
#df = df.rename(columns={'llm_answers_400': 'rag_answers_400'})

df["llm_answers_400"] = ""
#df["bleu_score_400"] = 0
df
```

```python
import nltk
for index, row in df.iterrows():
    #query = row["question"]
    reference = row["ground_truth"] # ground truth
    candidate_400 = row["rag_answers_400"]
    candidate_200 = row["rag_answers_200"]
    candidate_600 = row["rag_answers_600"]
    
    # Tokenize the reference and candidate responses
    reference_tokens = nltk.word_tokenize(reference.lower())
    candidate_tokens_400 = nltk.word_tokenize(candidate_400.lower())
    candidate_tokens_200 = nltk.word_tokenize(candidate_200.lower())
    candidate_tokens_600 = nltk.word_tokenize(candidate_600.lower())

    # Calculate BLEU score
    bleu_score_400 = nltk.translate.bleu_score.sentence_bleu([reference_tokens], candidate_tokens_400)
    bleu_score_200 = nltk.translate.bleu_score.sentence_bleu([reference_tokens], candidate_tokens_200)
    bleu_score_600 = nltk.translate.bleu_score.sentence_bleu([reference_tokens], candidate_tokens_600)
    #print("BLEU Score:", bleu_score)
    
    df.at[index,"bleu_score_400"] = bleu_score_400
    df.at[index,"bleu_score_200"] = bleu_score_200
    df.at[index,"bleu_score_600"] = bleu_score_600
#print("BLEU Score:", bleu_score)
```

### rouge score

```python
from rouge_score import rouge_scorer

# Example ground truth and LLM response
ground_truth = "The quick brown fox jumps over the lazy dog."
llm_response = "The quick red fox jumps over the lazy cat."

# Initialize the ROUGE scorer
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

# Calculate ROUGE scores
scores = scorer.score(ground_truth, llm_response)

# Print ROUGE scores
print("ROUGE-1:", scores['rouge1'])
print("ROUGE-2:", scores['rouge2'])
print("ROUGE-L:", scores['rougeL'])
```

```python
df["rouge1_precision_400"] = 0
df["rouge2_precision_400"] = 0
df["rougeL_precision_400"] = 0
df["rouge1_recall_400"] = 0
df["rouge2_recall_400"] = 0
df["rougeL_recall_400"] = 0
df["rouge1_fmeasure_400"] = 0
df["rouge2_fmeasure_400"] = 0
df["rougeL_fmeasure_400"] = 0
```

```python
from rouge_score import rouge_scorer

for index, row in df.iterrows():
    ground_truth = row["ground_truth"] # ground truth
    rag_response_400 = row["rag_answers_400"]
    rag_response_200 = row["rag_answers_200"]
    rag_response_600 = row["rag_answers_600"]
    
    scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    scores_400 = scorer.score(ground_truth, rag_response_400)
    scores_200 = scorer.score(ground_truth, rag_response_200)
    scores_600 = scorer.score(ground_truth, rag_response_600)
    
    #print("BLEU Score:", bleu_score)
    rouge1_precision_score_400 = scores_400['rouge1'].precision
    rouge1_precision_score_200 = scores_200['rouge1'].precision
    rouge1_precision_score_600 = scores_600['rouge1'].precision
    rouge2_precision_score_400 = scores_400['rouge2'].precision
    rouge2_precision_score_200 = scores_200['rouge2'].precision
    rouge2_precision_score_600 = scores_600['rouge2'].precision
    rougeL_precision_score_400 = scores_400['rougeL'].precision
    rougeL_precision_score_200 = scores_200['rougeL'].precision
    rougeL_precision_score_600 = scores_600['rougeL'].precision
    #print(rouge1_score)
    
    rouge1_recall_score_400 = scores_400['rouge1'].recall
    rouge1_recall_score_200 = scores_200['rouge1'].recall
    rouge1_recall_score_600 = scores_600['rouge1'].recall
    
    rouge2_recall_score_400 = scores_400['rouge2'].recall
    rouge2_recall_score_200 = scores_200['rouge2'].recall
    rouge2_recall_score_600 = scores_600['rouge2'].recall
    
    rougeL_recall_score_400 = scores_400['rougeL'].recall
    rougeL_recall_score_200 = scores_200['rougeL'].recall
    rougeL_recall_score_600 = scores_600['rougeL'].recall
    
    rouge1_fmeasure_score_400 = scores_400['rouge1'].fmeasure
    rouge1_fmeasure_score_200 = scores_200['rouge1'].fmeasure
    rouge1_fmeasure_score_600 = scores_600['rouge1'].fmeasure
    
    rouge2_fmeasure_score_400 = scores_400['rouge2'].fmeasure
    rouge2_fmeasure_score_200 = scores_200['rouge2'].fmeasure
    rouge2_fmeasure_score_600 = scores_600['rouge2'].fmeasure
    
    rougeL_fmeasure_score_400 = scores_400['rougeL'].fmeasure
    rougeL_fmeasure_score_200 = scores_200['rougeL'].fmeasure
    rougeL_fmeasure_score_600 = scores_600['rougeL'].fmeasure
    
    df.at[index,"rouge1_precision_400"] = rouge1_precision_score_400
    df.at[index,"rouge1_precision_200"] = rouge1_precision_score_200
    df.at[index,"rouge1_precision_600"] = rouge1_precision_score_600
    
    df.at[index,"rouge2_precision_400"] = rouge2_precision_score_400
    df.at[index,"rouge2_precision_200"] = rouge2_precision_score_200
    df.at[index,"rouge2_precision_600"] = rouge2_precision_score_600
    
    df.at[index,"rougeL_precision_400"] = rougeL_precision_score_400
    df.at[index,"rougeL_precision_200"] = rougeL_precision_score_200
    df.at[index,"rougeL_precision_600"] = rougeL_precision_score_600
    
    df.at[index,"rouge1_recall_400"] = rouge1_recall_score_400
    df.at[index,"rouge1_recall_200"] = rouge1_recall_score_200
    df.at[index,"rouge1_recall_600"] = rouge1_recall_score_600
    
    df.at[index,"rouge2_recall_400"] = rouge2_recall_score_400
    df.at[index,"rouge2_recall_200"] = rouge2_recall_score_200
    df.at[index,"rouge2_recall_600"] = rouge2_recall_score_600
    
    df.at[index,"rougeL_recall_400"] = rougeL_recall_score_400
    df.at[index,"rougeL_recall_200"] = rougeL_recall_score_200
    df.at[index,"rougeL_recall_600"] = rougeL_recall_score_600
    
    df.at[index,"rouge1_fmeasure_400"] = rouge1_fmeasure_score_400
    df.at[index,"rouge1_fmeasure_200"] = rouge1_fmeasure_score_200
    df.at[index,"rouge1_fmeasure_600"] = rouge1_fmeasure_score_600
    
    df.at[index,"rouge2_fmeasure_400"] = rouge2_fmeasure_score_400
    df.at[index,"rouge2_fmeasure_200"] = rouge2_fmeasure_score_200
    df.at[index,"rouge2_fmeasure_600"] = rouge2_fmeasure_score_600
    
    df.at[index,"rougeL_fmeasure_400"] = rougeL_fmeasure_score_400
    df.at[index,"rougeL_fmeasure_200"] = rougeL_fmeasure_score_200
    df.at[index,"rougeL_fmeasure_600"] = rougeL_fmeasure_score_600
    
    #print(rouge1_score)
```

### meteor score

```python
import nltk
from nltk.translate import meteor_score

# Sample LLM response and ground truth
llm_response = "The weather is sunny today."
ground_truth = "Today's weather is sunny."

# Tokenize the sentences
tokenized_llm_response = nltk.word_tokenize(llm_response)
tokenized_ground_truth = nltk.word_tokenize(ground_truth)

# Calculate the METEOR score
meteor_score_value = meteor_score.meteor_score([tokenized_llm_response], tokenized_ground_truth)

# Print the METEOR score
print("METEOR Score:", meteor_score_value)
```

```python
df["meteor_score_400"] = 0
```

```python
import nltk
for index, row in df.iterrows():
    
    ground_truth = row["ground_truth"] # ground truth
    rag_response_400 = row["rag_answers_400"]
    rag_response_200 = row["rag_answers_200"]
    rag_response_600 = row["rag_answers_600"]
    
    tokenized_rag_response_400 = nltk.word_tokenize(rag_response_400)
    tokenized_rag_response_200 = nltk.word_tokenize(rag_response_200)
    tokenized_rag_response_600 = nltk.word_tokenize(rag_response_600)
    
    tokenized_ground_truth = nltk.word_tokenize(ground_truth)

    meteor_score_value_400 = meteor_score.meteor_score([tokenized_rag_response_400], tokenized_ground_truth)
    meteor_score_value_200 = meteor_score.meteor_score([tokenized_rag_response_200], tokenized_ground_truth)
    meteor_score_value_600 = meteor_score.meteor_score([tokenized_rag_response_600], tokenized_ground_truth)
    
    #print("BLEU Score:", bleu_score)
    df.at[index,"meteor_score_400"] = meteor_score_value_400
    df.at[index,"meteor_score_200"] = meteor_score_value_200
    df.at[index,"meteor_score_600"] = meteor_score_value_600
    
    #print("BLEU Score:", bleu_score)
```
